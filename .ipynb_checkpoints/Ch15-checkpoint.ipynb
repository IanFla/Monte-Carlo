```python
def disks1(X):
    D1=(X[0]-1)**2+(X[1]-1)**2<=1
    D2=(X[0]-2.3)**2+(X[1]-2.3)**2<=0.27**2
    return D1|D2

A=np.array([[1,1],[1,-1]])/np.sqrt(2)
def disks2(Y):
    return disks1(np.linalg.inv(A).dot(Y))

X=uniform.rvs(size=10000,scale=3).reshape([2,-1])
flag1=disks1(X)
plt.plot(X[0][flag1],X[1][flag1],'.')

Y=uniform.rvs(size=10000,scale=3.5).reshape([2,-1])
Y[1]-=1
flag2=disks2(Y)
plt.plot(Y[0][flag2],Y[1][flag2],'.')
plt.show()
```

```python
cs=[np.sqrt(2),2.3*np.sqrt(2)]
def fcp1(y2):
    if np.abs(y2)>0.27:
        tmp=np.sqrt(1-y2**2)
        dy1=uniform.rvs(loc=cs[0]-tmp,scale=2*tmp)
    else:
        tmp=np.sqrt([1-y2**2,0.27**2-y2**2])
        ind=np.random.choice([0,1],p=tmp/tmp.sum())
        dy1=uniform.rvs(loc=cs[ind]-tmp[ind],scale=2*tmp[ind])
        
    return dy1
    
def fcp2(y1):
    if y1<2.5:
        tmp=np.sqrt(1-(y1-cs[0])**2)
    else:
        tmp=np.sqrt(0.27**2-(y1-cs[1])**2)
        
    dy2=uniform.rvs(scale=2*tmp)-tmp
    return dy2
```

```python
X0=np.array([1,1])
N=10000
Y=np.zeros([2,N+1])
Y[:,0]=A.dot(X0.reshape([2,1])).flatten()
for n in range(N):
    flag=np.random.choice([True,False])
    if flag:
        Y[1,n+1]=Y[1,n]
        Y[0,n+1]=fcp1(Y[1,n])
    else:
        Y[0,n+1]=Y[0,n]
        Y[1,n+1]=fcp2(Y[0,n])
```

```python
plt.plot(Y[0],Y[1],'.')
plt.plot(Y[0,0],Y[1,0],'go')
plt.plot(Y[0,N],Y[1,N],'ro')
plt.show()

plt.plot(Y[0,:101],Y[1,:101],'-')
plt.plot(Y[0,0],Y[1,0],'go')
plt.plot(Y[0,100],Y[1,100],'ro')
plt.show()

print(Y.mean(axis=1))

sb.histplot(Y[0])
plt.show()
```
