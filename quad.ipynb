```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.special import roots_jacobi, beta

import warnings
warnings.filterwarnings("ignore")
```

```python
def quadrature(a, b):
    x, w = roots_jacobi(1000, alpha=a, beta=b)
    f = lambda x: ((1 - x)**a) * ((1 + x)**b) / w.sum()
    fig, ax = plt.subplots(1, 3, figsize=(15, 4))
    
    ax[0].plot(x, f(x), label='density')
    ax[0].plot(x, f(x).max() * w/w.max(), label='weight')
    ax[0].hist(x, bins=50, density=True, weights=w, label='weighted points')
    
    ax[1].plot(x, f(x) / (w / w.mean()), label='point density')
    
    mask = (x >= -0.5) & (x <= 0.5)
    ax[2].plot(x[mask], (f(x) / (w / w.mean()))[mask], label='point density (zoomed)')
    
    ax[0].legend()
    ax[1].legend()
    ax[2].legend()
    plt.show()
```

```python
quadrature(0, 0) # Gauss-Legendre quadrature
```

```python
quadrature(0.5, 0.5) # Gauss-Chebyshev (second kind) quadrature
```

```python
quadrature(10, 10) # Gauss-Hermite quadrature
```

```python
quadrature(10, 0.5)
```

```python
quadrature(10, 0)
```
