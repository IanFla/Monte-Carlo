```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.special import roots_legendre, roots_jacobi, beta, roots_hermitenorm

import warnings
warnings.filterwarnings("ignore")
```

```python
f = lambda x: 0.5 * np.ones_like(x)
x, w = roots_legendre(1000)
plt.plot(x, f(x))
plt.plot(x, f(x).max() * w/w.max())
plt.hist(x, bins=50, density=True, weights=w)
plt.show()
plt.plot(x, f(x) / w)
plt.show()
```

```python
a = 1
b = 4
f = lambda x: ((1 - x)**a) * ((1 + x)**b) / ((2**(a + b + 1)) * beta(a+1, b+1))
x, w = roots_jacobi(1000, alpha=a, beta=b)
plt.plot(x, f(x))
plt.plot(x, f(x).max() * w/w.max())
plt.hist(x, bins=50, density=True, weights=w)
plt.show()
plt.plot(x, f(x) / w)
plt.show()
```

```python
f = lambda x: np.exp(-x**2/2) / np.sqrt(2*np.pi)
x, w = roots_hermitenorm(n=2000)
plt.plot(x, f(x))
plt.plot(x, f(x).max() * w/w.max())
plt.hist(x, bins=500, density=True, weights=w)
plt.xlim([-4, 4])
plt.show()
plt.plot(x, f(x) / w)
plt.xlim([-4, 4])
plt.show()
```
